### **Pandas 的使用**

#### **匯入 Pandas 套件**
使用 `import PackageName as abbr` 的方式匯入 `pandas`，像是
```python
import pandas as pd
```
這句話，來將 `pandas` 匯入，並簡稱為 `pd`。

````python
import pandas as pd
````

#### **使用 pandas 讀取CSV資料**
使用 `read_csv()` function 來讀取 CSV 檔案，並建立 Pandas 的 dataframe。

````python
# 設定路徑
fpath = 'input/breast-cancer-wisconsin-data/data.csv'

# 將資料讀取並賦值給 dataframe
df = pd.read_csv(fpath)
````

````python
# 顯示 dataframe 內容
df
````

````python
# 使用 type() 查看資料類別，會顯示是 dataframe
type(df)
````

#### **顯示資料的前幾筆 (預覽)**
使用 Dataframe 物件的 `head()` method 來 return Dataframe 的前N筆資料。 (預設是 5 筆)

````python
df.head() # 根據位置返回對象的前5行
````

````python
df.head(10) # 根據位置返回對象的前10行
````

#### **查看 Dataframe 物件的資訊**
使用 `info()` method 來取得 Dataframe 物件的資訊

````python
df.info()
````

#### **刪除 Dataframe 欄位**
使用 `drop()` method 來刪除 Dataframe 的欄或列。
```python
df.drop(['Column_Name'], axis = 1, inplace = True)
```

````python
# axis = 1 代表刪除 column，等於 0 代表刪除 row
df.drop(['Unnamed: 32'], axis = 1, inplace = True)

# 可以注意叫做 Unnamed: 32 的 column 已經被刪掉了
df.info()
````

##### **取出 Dataframe 特定欄位的值**
使用 `df[['ColumnName']].value` 取出特定欄位的值

````python
df[['compactness_worst', 'symmetry_worst']].values
````

#### **取出 Dataframe 中所有欄位名稱**
使用 Dataframe 的 `columns` 屬性來取得所有欄位名稱。

````python
df.columns
````
